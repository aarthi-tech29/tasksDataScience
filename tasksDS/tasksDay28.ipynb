##### 1. import pandas as pd

##### 3. from scipy import stats 

```python
import pandas as pd
import numpy as np
from scipy import stats
```

##### 4. df = pd.DataFrame({'cat1': ['A', 'A', 'B'], 'cat2': ['X', 'Y', 'X']})

```python
# Create sample data for Chi-Square test
df = pd.DataFrame({
    'cat1': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'cat2': ['X', 'Y', 'X', 'Y', 'X', 'X', 'Y', 'Y', 'X', 'Y']
})
print(df)
```

##### 5. contingency_table = pd.crosstab(df['cat1'], df['cat2']) 

```python
# Create contingency table
contingency_table = pd.crosstab(df['cat1'], df['cat2'])
print("Contingency Table:\n", contingency_table)
```

##### 6. chi2, p_val, dof, expected = stats.chi2_contingency(contingency_table)

```python
# Run Chi-Square Test
chi2, p_chi, dof, expected = stats.chi2_contingency(contingency_table)
```

##### 7. if p_val < 0.05: print("Dependent") 

```python
# Check dependence
if p_chi < 0.05:
    print("Chi-Square Conclusion: Dependent")
else:
    print("Chi-Square Conclusion: Independent")
```

##### 8. group1 = np.random.normal(10, 2, 30)
##### 9. group2 = np.random.normal(12, 2, 30) 
##### 10. group3 = np.random.normal(10.5, 2, 30) 

```python
# Generate data for ANOVA
np.random.seed(42)
group1 = np.random.normal(10, 2, 30)
group2 = np.random.normal(12, 2, 30)
group3 = np.random.normal(10.5, 2, 30)
```

##### 11. f_stat, p_val = stats.f_oneway(group1, group2, group3)

```python
# Run ANOVA
f_stat, p_anova = stats.f_oneway(group1, group2, group3)
```

##### 12. if p_val < 0.05: print("Means are different") 

```python
# Check if means are different
if p_anova < 0.05:
    print("ANOVA Conclusion: Means are different")
else:
    print("ANOVA Conclusion: Means are not different")
```

##### 13. print(f"Chi-Square p-value: {p_val:.4f}") 

```python
print(f"Chi-Square p-value: {p_chi:.4f}")
print(f"ANOVA p-value: {p_anova:.4f}")
```
