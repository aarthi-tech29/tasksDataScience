##### 8. sample_means = [np.mean(np.random.choice(population, 30)) for _ in range(1000)]

```python
sample_means = [np.mean(np.random.choice(population, 30)) for _ in range(1000)]
print(sample_means)

# Generate 1,000 sample means, each from a sample of size 30
# sample_means - simulation of sampling distribution of means
```

##### 9. df.groupby('category').apply(lambda x: x.sample(n=10))

```python
df['category'] = np.random.choice(['A', 'B', 'C', 'D'], size=len(df))
# print(df['category'])
sample_by_category = df.groupby('category').apply(lambda x: x.sample(n=10)) # 10 samples per category
# print(sample_by_category)
print("\nRandom 10 samples per category:\n", sample_by_category.head(20))


# Take 10 random samples from each category
# ['A', 'B', 'C', 'D'] - This is a list of possible categories.
# size=len(df) - The number of elements to generate is equal to the number of rows in df.
# np.random.choice() - Randomly picks one element from the list for each row.
```

##### 10. from scipy.stats import norm

```python
from scipy.stats import norm
# norm - normal distribution
```

##### 11. z_score = (sample.mean() - population.mean()) / (population.std() / np.sqrt(len(sample)))

```python
z_score = (sample.mean() - population.mean()) / (population.std() / np.sqrt(len(sample)))

print("Sample of 100 values mean:", sample.mean())
print("Population mean:", population.mean())
print("Z-score:", z_score)
# z-score - How many standard deviations is the sample mean away from the population mean
```

##### 12. p_value = norm.cdf(z_score)

```python
p_value = norm.cdf(z_score)
print("P-value:", p_value)
# Z-score and p-value calculation for hypothesis testing
# norm.cdf()- normal distribution, Cumulative Distribution Function(cdf)
# the probability that a value is less than or equal to a given Z-score
# 54% chance sample mean is normal, 46% chance it's unusual, Sample mean is similar to population mean
# p < 0.05 -Reject H₀-sample is significantly different
# p > 0.05 - Fail to reject H₀-sample is normal/common
# Reject H₀ = Something is happening
# Fail to reject H₀ = Nothing new. No change detected.
```

##### 13. standard_error = np.std(sample_means) 

```python
standard_error = np.std(sample_means)
print("Standard error of sample means:", standard_error)


# Calculate standard error of the sample means
# Standard Error (SE) - How much your sample mean is expected to vary from the true population mean, It measures the accuracy of your sample mean
# It shows how stable or unstable your sample mean is when you repeat sampling many times
# SE = 0.05023 sample means vary very little from the true mean
```
